```python
data_brut = pd.read_excel('coraux-substrat_Serie_Tiahura_87-2023.xlsx', sheet_name='Brut')
```

```python
# extract the season column
Season = data_brut['Season']
data_brut.drop(columns=['Season'], inplace=True)
```

```python
# convert date to the first day of the month
data_brut['Date'] = data_brut['Date'].dt.strftime('%Y-%m')
data_brut = data_brut.rename(columns={'Date': 'YearSeason'})
data_brut.drop(columns=['Year'], inplace=True)
```

```python
data_brut.drop(columns=['Observer', ], inplace=True)
```

```python
data_brut['Substrate'].unique()
```

```python
data_brut.drop(columns=['Remarques'], inplace=True)
```

```python
data_brut['Habitat'].unique()
```

```python
data_brut_frin = data_brut[data_brut['Habitat'] == 'Fringing reef']
data_brut_for = data_brut[data_brut['Habitat'] == 'Fore reef']
data_brut_cha = data_brut[data_brut['Habitat'] == 'Channel']
data_brut_alg = data_brut[data_brut['Habitat'] == 'Algal Crest']
data_brut_bar = data_brut[data_brut['Habitat'] == 'Barrier reef']

data_brut_frin.drop(columns=['Habitat'], inplace=True)
data_brut_for.drop(columns=['Habitat'], inplace=True)
data_brut_cha.drop(columns=['Habitat'], inplace=True)
data_brut_alg.drop(columns=['Habitat'], inplace=True)
data_brut_bar.drop(columns=['Habitat'], inplace=True)
```

```python
# def a function that does the same thing as above
def count_normalize(data):
    data = data.groupby(['YearSeason', 'Substrate']).size().reset_index(name='counts')
    data = data.pivot(index='YearSeason', columns='Substrate', values='counts')
    data = data.fillna(0)
    data = data.div(data.sum(axis=1), axis=0)
    return data

```

```python
data_brut_frin = count_normalize(data_brut_frin)
data_brut_for = count_normalize(data_brut_for)
data_brut_cha = count_normalize(data_brut_cha)
data_brut_alg = count_normalize(data_brut_alg)
data_brut_bar = count_normalize(data_brut_bar)
```

```python
dict_substrate = {}
dict_substrate['A'] = 'Algae'
dict_substrate['B'] = 'Others'
dict_substrate['C'] = 'Live Coral'
dict_substrate['D'] = 'Pavement'
dict_substrate['O'] = 'Others'
dict_substrate['P'] = 'Calcified Algae'
dict_substrate['M'] = 'Others'
dict_substrate['R'] = 'Rubble'
dict_substrate['S'] = 'Sand'
dict_substrate['T'] = 'Turf'
dict_substrate['?'] = 'Others'
```

```python
# replace column names with the dictionary
data_brut_frin = data_brut_frin.rename(columns=dict_substrate)
data_brut_for = data_brut_for.rename(columns=dict_substrate)
data_brut_bar = data_brut_bar.rename(columns=dict_substrate)
```

```python
# add together the columns with the same name
data_brut_frin = data_brut_frin.groupby(data_brut_frin.columns, axis=1).sum()
data_brut_for = data_brut_for.groupby(data_brut_for.columns, axis=1).sum()
data_brut_bar = data_brut_cha.groupby(data_brut_bar.columns, axis=1).sum()

```
