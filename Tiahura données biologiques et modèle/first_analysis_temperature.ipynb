```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
```

```python
temp_houlo = pd.read_csv('SBE26_houlo/temprature-c-tiahura.csv', sep = ';')
temp_multi = pd.read_csv('SBE16_multi/temprature-c-tiahura.csv', sep = ';')
```

```python
temp_houlo = temp_houlo.set_index('DateTime')
temp_multi = temp_multi.set_index('DateTime')
```

```python
temp_houlo.rename(columns={'°C' : 'T_houlo'}, inplace=True)
```

```python
temp_multi.rename(columns={'°C' : 'T_multi'}, inplace=True)
```

```python
temps = pd.merge(temp_houlo, temp_multi, left_index=True, right_index=True, how = 'outer')
```

```python
# Replace commas with dots and convert to float
temps['T_houlo'] = temps['T_houlo'].str.replace(',', '.').astype(float)
temps['T_multi'] = temps['T_multi'].str.replace(',', '.').astype(float)
```

```python
plt.figure(figsize=(25, 12))
plt.plot(temps.index, temps['T_houlo'])
plt.plot(temps.index, temps['T_multi'])

# Add labels and a legend
plt.xlabel('Date and Time')
plt.ylabel('Temperature')
plt.title('Temperature over Time')
plt.legend()
plt.tight_layout()
```

```python
plt.figure(figsize=(25, 12))
plt.plot(temps.index, temps['T_houlo'] - temps['T_multi'])

# Add labels and a legend
plt.xlabel('Date and Time')
plt.ylabel('Temperature Difference')
plt.title('Temperature Difference over Time')
plt.tight_layout()
```

### on merge T_houlo et p35

```python
# merge T_houlo and T_p35 following these rules: if T_houlo is NaN but T_p35 is not, then T_merged = T_p35, and vice versa, if both are not NaN, then T_merged = (T_houlo + T_p35) / 2, if both are NaN, then T_merged = NaN

def merge_temps(T1, T2):
    T_merged = T1.copy()
    T_merged[T1.isnull() & T2.notnull()] = T2
    T_merged[T1.notnull() & T2.isnull()] = T1
    T_merged[T1.notnull() & T2.notnull()] = (T1 + T2) / 2
    return T_merged

temps_merged = merge_temps(temps['T_houlo'], temps['T_p35'])

plt.figure(figsize=(25, 12))
plt.plot(temps_merged.index, temps_merged, label = 'T_merged')
plt.plot(temps.index, temps['T_houlo'], label = 'T_houlo')
plt.plot(temps.index, temps['T_p35'], label = 'T_p35')
plt.title('Merged Temperature between Houlo and P35')
plt.xlabel('Date and Time')
plt.ylabel('Temperature')
plt.legend()
plt.tight_layout()
```

```python
# same work with T_multi and T_p25
temps_merged_2 = merge_temps(temps['T_multi'], temps['T_p25'])

plt.figure(figsize=(25, 12))
plt.plot(temps_merged.index, temps_merged, label = 'T_merged')
plt.plot(temps.index, temps['T_multi'], label = 'T_multi')
plt.plot(temps.index, temps['T_p25'], label = 'T_p25')
plt.title('Merged Temperature between Multi and P25')
plt.xlabel('Date and Time')
plt.ylabel('Temperature')
plt.legend()
plt.tight_layout()
```
